```python
import skimage.io as io
import matplotlib.pyplot as plt

img = io.imread('data/punto10_spectrum.png')
img.shape
```

El ancho de la imagen original es 556 pixeles

```python
plt.imshow(img)
plt.show()
```

Se recorta para que quede en el rango [0, 80], quedando con un ancho de 492 pixeles:

```python
# recorte hasta desde 0 hasta 80
img = img[:, 1:493, :]
print(img.shape)
plt.imshow(img)
plt.show()
```

#### Prueba de los valores de las lneas de emisin de **Au**

Tomamos los valores de energa de las lneas del oro de la tabla del Lawrence Berkeley National Laboratory:



```python
# x = valor_energia * 492 / 80

x1 = 9.7 * 492 / 80 # L_alpha_1

x2 = 11.4 * 492 / 80 # L_beta_1

x3 = 68.8 * 492 / 80 # K_alpha_1

x4 = 78 * 492 / 80 # K_beta_1
```

```python
plt.imshow(img)
plt.title('Lneas de emisin del Au')
plt.axvline(x1, color='red', linestyle='--', label=r"$L\alpha_1$")
plt.axvline(x2, color='maroon', linestyle='--', label=r"$L\beta_1$")
plt.axvline(x3, color='lime', linestyle='--', label=r"$K\alpha_1$")
plt.axvline(x4, color='darkgreen', linestyle='--', label=r"$K\beta_1$")
plt.legend(loc='upper left')
plt.axis('off')
plt.show()
```

Encontramos que los valores de lneas de emisin del oro (Au) de la tabla son los que mejor coinciden con los picos del espectro dado, por lo que logramos identificar que el elemento desconocido es Au.
