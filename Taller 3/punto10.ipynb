```python
import skimage.io as io
import matplotlib.pyplot as plt

img = io.imread('data/10_spectrum.png')
img.shape
```

El ancho de la imagen original es 552 pixeles

```python
plt.imshow(img)
plt.show()
```

Se recorta para que quede en el rango [0, 80], quedando con un ancho de 434 pixeles:

```python
# recorte hasta desde 0 hasta 80
img = img[:, 59:493, :]
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
ancho_pixeles = img.shape[1]
ancho_energias = 80
```

#### Prueba de los valores de las líneas de emisión de **Au**

Tomamos los valores de energía de las líneas del oro de la tabla del Lawrence Berkeley National Laboratory:



```python
# de la tabla obtenemos los valores de energías
L_alpha_1_Au = 9.7133

L_beta_1_Au = 11.4423

K_alpha_1_Au = 68.8037

K_beta_1_Au = 77.984

# convertimos las energias a las posiciones en el grafico
x1 = L_alpha_1_Au * ancho_pixeles / ancho_energias 

x2 = L_beta_1_Au * ancho_pixeles / ancho_energias 

x3 = K_alpha_1_Au * ancho_pixeles / ancho_energias 

x4 = K_beta_1_Au * ancho_pixeles / ancho_energias 
```

```python
plt.imshow(img)
plt.title('Líneas de emisión del Au')
plt.axvline(x1, color='red', linestyle='--', label=r"$L\alpha_1$")
plt.axvline(x2, color='maroon', linestyle='--', label=r"$L\beta_1$")
plt.axvline(x3, color='lime', linestyle='--', label=r"$K\alpha_1$")
plt.axvline(x4, color='darkgreen', linestyle='--', label=r"$K\beta_1$")
plt.legend(loc='upper left')
plt.axis('off')
plt.show()
```

Encontramos que los valores de líneas de emisión del oro (Au) de la tabla son los que mejor coinciden con los picos del espectro dado, por lo que logramos identificar que el elemento desconocido es Au. 

### (b)

Revisamos las relaciones de intensidades entre los picos:

```python
img = io.imread('data/10_spectrum.png')
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
img = img[69:408, :, :]
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
alto_pixeles = img.shape[0]
alto_counts = 150
```

```python
intensidad_L_alpha_1_Au = 40
y1 = (alto_counts-intensidad_L_alpha_1_Au-1) * alto_pixeles / alto_counts

intensidad_L_beta_1_Au = 30
y2 = (alto_counts-intensidad_L_beta_1_Au) * alto_pixeles / alto_counts

intensidad_K_alpha_1_Au = 155
y3 = (alto_counts-intensidad_K_alpha_1_Au) * alto_pixeles / alto_counts

intensidad_K_beta_1_Au = 40
y4 = (alto_counts-intensidad_K_beta_1_Au+1) * alto_pixeles / alto_counts
```

```python
plt.imshow(img)
plt.title('Intensidades observadas de picos del Au')
plt.axhline(y1, color='red', linestyle='--', label=r"$L\alpha_1$")
plt.axhline(y2, color='maroon', linestyle='--', label=r"$L\beta_1$")
plt.axhline(y3, color='lime', linestyle='--', label=r"$K\alpha_1$")
plt.axhline(y4, color='darkgreen', linestyle='--', label=r"$K\beta_1$")
plt.legend(loc='upper center')
plt.axis('off')
plt.show()
```

Debido a la mala posición de la leyenda en el gráfico original, no es posible calcular con precisión la intensidad del pico $K\beta_1$. La estimamos en 155. Las intensidades encontradas fueron:

Intensidad $L\alpha_1$ = 40

Intensidad $L\beta_1$ = 30

Intensidad $K\alpha_1$ = 155

Intensidad $K\beta_1$ = 40


Procedemos a calcular la relación de las intensidades entre los picos $K\alpha_1$/$K\beta_1$ y $L\alpha_1$/$L\beta_1$:

```python
relacion_K = intensidad_K_alpha_1_Au / intensidad_K_beta_1_Au
relacion_L = intensidad_L_alpha_1_Au / intensidad_L_beta_1_Au

print(f"Relación K: {round(relacion_K,2)}")
# se aproxima a dos decimales 
    
print(f"Relación L: {round(relacion_L,2)}")
```

La relación $K\alpha_1$/$K\beta_1$ encontrada fue de 3.88, por lo que se encuentra en el rango esperado para esta relación de intensidades. Además, se comprueba que los picos de la serie L son menos intensos que los de la serie K.
